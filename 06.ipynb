```python
import pandas as pd
mpg = pd.read_csv('mpg.csv')
mpg
```

```python
# q02
mpg_audi = mpg.query('manufacturer == "audi"')
mpg_toyota = mpg.query('manufacturer == "toyota"')

print(f'audi의 도시연비 평균 : {mpg_audi['cty'].mean():.2f}')
print(f'toyota의 도시연비 평균 : {mpg_toyota['cty'].mean():.2f}')
if mpg_audi['cty'].mean() > mpg_toyota['cty'].mean():
    print('audi의 평균연비가 높습니다')
else:
    print('toyota의 평균연비가 높습니다')
```

```python
# q03
mpg_cfh = mpg.query('manufacturer in ["chevrolet","ford","honda"]')

ave_hwy = mpg_cfh['hwy'].mean()
print(f'쉐보레, 포드, 혼다의 고속도로 평균연비는 {ave_hwy:.2f}')
```

```python
# groupby() - 전체가 아닌 그룹별 , agg()- 요약해주는 함수. agg(변수 = ('키','함수명'))
import pandas as pd
exam = pd.read_csv('exam.csv')
#exam.agg(mean_math = ('math','mean'))
exam.groupby('nclass',as_index=False).agg(mean_math=('math','mean'))
```

```python
exam.groupby('nclass').agg(수학평균=('math','mean'),
                           합계=('math','sum'),
                           중간=('math','median'),
                           인원 = ('math','count'))                          
```

```python
exam.groupby('nclass').agg(mean_math=('math','mean'),
                          std_math=('math','std'),
                          median_math=('math','median'),
                          min_math=('math','min'),
                          max_math=('math','max'),
                          count_math=('math','count'))
```
