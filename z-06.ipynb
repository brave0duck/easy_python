## 06-3. 데이터 추출

```python
import pandas as pd
import numpy as np

mpg = pd.read_csv('mpg.csv')
mpg_mod = mpg.copy()
mpg_mod
```

```python
# 혼자서 해보기 
# Q1. mpg데이터에서 category,cty를 추출해 새 변수에 대입. 새 변수의 일부를 출력해서 확인
new_df = mpg_mod[['category','cty']]
new_df.head()

# Q2. 종류에 따른 연비체크. 카테고리 'suv','compact' 자동차중 어떤 차의 도시연비 평균이 더 좋을까?
suv_mean = mpg_mod.query("category in ['suv']").agg(cty_mean = ('cty','mean')).iloc[0,0]
compact_mean = mpg_mod.query("category in ['compact']").agg(cty_mnea = ('cty','mean')).iloc[0,0]

if suv_mean >= compact_mean:
    print(f'suv가 더 연비가 높습니다. suv:{suv_mean:.2f}, compact:{compact_mean:.2f}')
else:
    print(f'compact 가 더 연비가 높습니다. compact:{compact_mean:.2f}, suv:{suv_mean:.2f}')
```

## 06-5. 파생변수 추가하기

```python
import pandas as pd
import numpy as np

exam = pd.read_csv('exam.csv')
exam = exam.assign(total = exam['math'] + exam['english']+ exam['science'])     # 추가
exam = exam.assign(test = np.where(exam['science'] >= 60,'pass','fail'))        # numpy를 이용해 추가할수있다
exam = exam.assign(test_mean = lambda x: x['total']/3)                          # lambda함수를 쓰는방법도 있다
exam
```

```python
# 혼자서 해보기
# Q1. mpg복사본을 만들고 cty와 hwy를 더한 합산연비를 추가
mpg_copy = mpg.copy()
mpg_copy = mpg_copy.assign(total = lambda x : x['cty']+x['hwy'])

# # Q2. 앞에서 합산연비를 2로 나눠 평균연비변수 추가
mpg_copy = mpg_copy.assign(total_mean = lambda x : x['total'] /2 )

# # Q3. 평균연비가 가장 높은 자동차 3종의 데이터 출력
print(mpg_copy.sort_values('total_mean',ascending=False).head(3))   #폭스바겐 비틀, 제타타

# Q4. 1-3번 문제를 하나로 연결한 판다스구문을 만들어서 실행해보라. mpg원본사용
mpg = mpg.assign(total = lambda x: x['cty']+ x['hwy'], total_mean = lambda x: x['total']/2).sort_values('total_mean',ascending=False).head(3)
```
