```python
import pandas as pd
mpg = pd.read_csv('mpg.csv')
mpg
```

```python
# 아우디 차량중 구동방식 기준으로 몇대의 차가 있는가?
mpg.query('manufacturer == "audi"').groupby(['drv']).agg(n=('drv','count'))
# 쉐보레 차량중 구동방식 기준으로 몇대의 차가 있는가?
mpg.query('manufacturer == "chevrolet"').groupby(['drv']).agg(n = ('drv','count'))

# <예제>
# 1. 제조회사별로, 'suv'자동차의 도시고속도로 합산연비 평균을 구해, 내림차순으로 정렬하고, 1-5위까지 출력하기
mpg.query('category == "suv"')\
    .assign(suv = (mpg['hwy'] + mpg['cty']) / 2)\
    .groupby(['manufacturer'])\
    .agg(mean_suv = ('suv','mean'))\
    .sort_values('mean_suv',ascending=False)\
    .head(10)
```

```python
# 166p . mpg데이터를 이용해 분석문제를 해결
# q1,q2 어떤 차종의 도시 연비가 높은지 비교. 카테고리별 cty평균을 구하라
mpg.groupby('category',as_index=False).agg(cty_mean = ('cty','mean')).sort_values('cty_mean',ascending=False)
```

```python
# q3 . 어떤회사의 hwy가 가장 높은지 알고싶다. hwy평균이 가장 높은 회사 3곳 출력
mpg.groupby('manufacturer',as_index=False).agg(hwy_mean = ('hwy','mean')).sort_values('hwy_mean',ascending=False).head(3)
```

```python
# q4.어느 회사에서 'compact'차종을 가장 많이 생산하는지 알고싶다. 회사별 'compact' 차종수를 내림차순으로 정렬,출력
mpg.query('category == "compact"').groupby('manufacturer').agg(count= ('manufacturer','count')).sort_values('count',ascending=False)
```

```python
fuel = pd.DataFrame({'fl':['c','d','e','p','r'],
             'price_fl': [2.35,2.38,2.11,2.76,2.22]})

new_mpg = pd.merge(mpg,fuel, how='left', on='fl')
new_mpg[['model','fl','price_fl']].head(20)
```
